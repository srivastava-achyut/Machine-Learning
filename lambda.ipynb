```python
def addition(a,b):
    return a+b

addition(1,3)
```

```python
addition=lambda a,b:a+b
type(addition)
print(addition(5,6))
```

```python
def even(num):
    if num%2==0:
        return True

even(24)
```

```python
even1=lambda num:num%2==0
even(12)
```
