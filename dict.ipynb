Dictionary

```python
##creating a dictionary

##it has a key value pair

empty_dict={}
print(type(empty_dict))
```

```python
student={"name":"Krish","age":32,"grade":24}
print(student)
```

```python
## accessing dictionary element
stu={"name":"Achyut","age":19,"grade":'A'}
print(stu)
print(stu['age']) #19
```

```python
##accessing using get method

print(stu.get('grade'))
print(stu.get('last_name','not Available'))
```

```python
## modifying dictionary element
## dictionary are mutable so we add and delete element

student["age"]=45  ##update the key
print(student)
student["address"]="uttarpradesh"
print(student)
```

```python
##delete the key

del student["grade"]

print(student)

```

```python
keys=student.keys()  ##help to get all the keys 
print(keys)
values=student.values()  ##help to get all the values
print(values)
items=student.items()  ##get all the key value
print(items)
```

```python
##shallow copy

student_copy=student
print(student_copy)
```

```python
student["name"]="Krish1"
print(student_copy)
print(student)
```

```python
student_copy1=student.copy() ## shallow copy
print(student_copy1)
print(student)
```

```python
student["name"]="Krish@"
print(student_copy1)
print(student)
```

```python
## iterating over loops

for keys in student.keys():
    print(keys)
```

```python
##nested dictionary

students={
  "student1":{"name":"Achyut","age":19,"grade":'A'},
  "student2":{"name":"Abhishek","age":20,"grade":'A'},
  "student3":{"name":"Satish","age":21,"grade":'C'}


}
print(students)
```

```python
##access nested dict elements

print(students["student2"]["name"])
print(students["student3"]["age"])
```

```python
##Dictionary Comprehension

squares={x:x**2 for x in range(5)}
print(squares)
```

```python
### conditional dictonary comprehension

evens={x:x**2 for x in range(10) if x%2==0}
print(evens)
```

```python
## practical examples

##use a dictionary to count the frequency of elements in list

number=[1,2,2,3,3,4,4,4,4,4]
frequency={}

for numbers in number:
    if numbers in frequency:
        frequency[numbers]+=1
    else:
        frequency[numbers]=1
print(frequency)
```
