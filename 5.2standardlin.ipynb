```python
##csv

import csv
with open('example.csv',mode='w',newline='') as file:
   write=  csv.writer(file)
   write.writerow(['name','age'])
   write.writerow(['Krish',23])

with open('example.csv',mode='r') as file:
    reader=csv.reader(file)
    for row in reader:
        print(row)
   
```
