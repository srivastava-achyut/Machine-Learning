Data Visualization With Matplotlib

Matplotlib is a powerful plotting library for Python that enables the creation of static, animated, and intractive visualization.
It is widely used for data visualization in data science and analytics. In this lesson, we will cover the basics of Matplotlib, including creating various types of plots and customizing them.

```python
import matplotlib.pyplot as plt

```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

## create aline plot
plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Basic Line Plot')
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.plot(x,y,color='red',linestyle='--',marker='o',linewidth=3,markersize=9)
plt.grid(True)
```

```python
## Multiple Plots
## Sample Data

x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,5]
plt.figure(figsize=(9,5))
plt.subplot(2,2,1)
plt.plot(x,y1,color='green')
plt.title("Plot 1")

plt.subplot(2,2,2)
plt.plot(x,y2,color='green')
plt.title("Plot 2")

plt.subplot(2,2,3)
plt.plot(y1,x,color='green')
plt.title("Plot 3")
```

```python
## Bar plor

categories=['A','B','C','D','E']
values=[5,7,3,8,6]

##create a bar plot

plt.bar(categories,values,color='blue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```

HISTOGRAM


```python
data=[1,2,3,1,3,4,2,4,5,3,2,1,4,5,1,1,1,2,2,2,3,4,3,3,3,4,4,5,5,5]

#Create a histogram
plt.hist(data,bins=5,color='orange',edgecolor='black')
```

```python
## scatter plot

#Sample data

x=[1,2,3,4,5]
y=[2,3,4,5,6]

plt.scatter(x,y,color="blue",marker='x')
```

```python
## pie chart

labels=['A','B','C','D','E']
sizes=[30,40,10,20,50]
colors=['gold','yellowgreen','lightcoral','blue','lightsky']
explode=(0.2,0,0,0,0)

plt.pie(sizes,explode=explode,labels=labels,autopct='%1.1f%%',shadow=True)
```

```python
#sales data visualization

import pandas as pd
sales_data=pd.read_csv('sales.csv')
sales_data.head(5)
```

```python
## plot total sales by product
total_sales_by_product=sales_data.groupby('Product')['Sales'].sum()
print(total_sales_by_product)
```

```python
total_sales_by_product.plot(kind='bar',color='teal')
```

```python
##sales trend over time

sales_trend=sales_data.groupby('OrderDate')['Sales'].sum().reset_index()
plt.plot(sales_trend['OrderDate'],sales_trend['Sales'])
```
