Tuples are similar like list but they are immutable means we can not change them after decleraton

```python
## creating a tuple

empty_tuple=()
print(empty_tuple)  ##()
print(type(empty_tuple))

```

```python
numbers=(1,2,3,4)
print(numbers)
```

```python
list=[1,2,3,4]
numbers=tuple(list)
print(numbers)
```

```python
mixed_tuple=(1,"hello",4.567)
print(mixed_tuple)
```

```python
##Accesing tuple elements 

numbers[0]
```

```python
numbers
```

```python

### Tuple operation

numbers+mixed_tuple ## concatination

```

```python
##immutable nature of tuples

list=[1,2,3,4]
print(list)

list[0]="Krish"
print(list)
```

```python
values=(1,2,3,4,5)
print(values)

values[0]="Krish" ##shows error because we can not change the tuple element
print(values)
```

```python
##tuple methods

numbers
print(numbers.count(0))
```

```python
print(numbers.index(3))
```

```python
## packing and unpacking tuple

packed_tuple=1,"hello",3,24
print(packed_tuple)
```

```python
##unpacking

packed_tuple = (10, 20, 30)

a, b, c = packed_tuple
print(a)
print(b)
print(c)
```

```python
numbers=(1,2,3,4,5,6)
first,*middle,last=numbers
print(first)  ###  1

print(middle)   ### [2,3,4,5]
print(last)    ## 6
```

```python
###nested tuple
## nested list
list=[[[],1,2,],5,6,6,7]
print(list)
print(list[0])
print(list[0][0])
```

```python
## nested tuple  

nested_tuple=((1,2,3),("apple","mango","banana"),(3,45,True,False))

print(nested_tuple)
print(nested_tuple[0])
print(nested_tuple[0][2])
```
