File Opearation: Read and Write File

```python
##Read and write

with open('example.txt','r') as file:
    content=file.read()
    print(content)
    
```

```python
#read line by line

with open('example.txt','r') as file:
    for line in file:
        print(line.strip())
```

```python
##writing a file(overwriting)

with open('example.txt','w') as file:
    file.write('Hello World\n')
    file.write('this is new line')
```

```python
## write a file without overwriting

with open('example.txt','a') as file:
    file.write("Append opearation taking place")  
```

```python
###writing a list of lines to a file
lines=['First line\n','Second line\n','third line\n']
with open('example.txt','a') as file:
    file.writelines(lines)
    
```

```python
##binary files

data=b'\x00\x01\x02\x03\x04'
with open('example.bin','wb') as file:
    file.write(data)
```

```python
##Reading a binary file

with open('example.bin','rb') as file:
    content=file.read()
    print(content)
```

```python
##Read the content from source text file and write to destination text file
with open('example.txt','r') as source_data:
    content=source_data.read()

with open('destination.txt','w') as dest_data:
    dest_data.write(content)
    
```

```python
## writing and then reading a file

with open('example.txt','w+') as file:
    file.write('Hello World\n')
    file.write('This is a new line\n')

    ## move the file cursor to the begining
    #file.seek(0)

    #Read the content of the line
    content=file.read()
    print(content)


```

```python
## joining paths

dir_name="folder"
file_name='file.txt'
full_path=os.path.join(dir_name,file_name)
print(full_path)
```

```python
path='example.txt'
if os.path.exists(path):
    print(f"The path '{path}' exists")

else:
    print(f"the path '{path}' does not exist")
```
