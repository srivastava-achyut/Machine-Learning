PANDAS-DATAFRAME AND SERIES 
Pandas is a powerful data manipulation library inPython, widely used for data analysis and data cleaning

```python
import pandas as pd
```

```python
##series

import pandas as pd
data=[1,2,3,4,5,6]
series=pd.Series(data)
print(series)
```

```python
## Create a Series from dictionary

data={'a':1,'b':2,'c':3,'d':4}
series_dict=pd.Series(data)
print(series_dict)
```

```python
data=[10,20,30]
index=['a','b','c']
pd.Series(data,index=index)

```

```python
## Dataframe
## create a dataframe

data={
    'Name':['Krish','John','Jack'],
    'Age':[25,30,45],
    'City':['Banglore','New York','Texas']
}
df=pd.DataFrame(data)
print(df)
print(type(df))
```

```python
import numpy as np
np.array(df)
```

```python
## Create a dataframe from a list of dictionary

data=[
   {'Name':'Achyut','Age':32,'City':'Banglore'},
   {'Name':'Srikant','Age':35,'City':'Newyork'},
   {'Name':'Ajay','Age':34,'City':'Jaipur'},

]
df=pd.DataFrame(data)
print(df)
print(type(df))

```
