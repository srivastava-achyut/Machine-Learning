FUNCTION 

```python
##FUNCtion is block of code that performs specific task

#function for even odd

def even_odd(num):
    """This function find even or odd"""
    if num%2==0:
        print(num ,"is even")
    else:
        print(num ,"is odd")
even_odd(5)
```

```python
## function with multiple parameters


def add(a,b):
    return a+b

add(4,5)
```

```python
##default parameters

def greet(name):
    print(f"Hello {name} Welcom to the Paradise")

greet("Achyut")
```

```python
## variable length argument

##positional and keyword arguments

def print_numbers(*args):
    for numbers in args:
        print(numbers)
```

```python
print_numbers(1,2,3,4,5,6,7)
```

```python
### keyword Arguments

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(name="krish",age="32",grade='A')
```

```python
### return statement

def multiply(a,b):
    return a*b
multiply(2,3)
```
