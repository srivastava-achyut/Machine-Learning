Celcius to Ferhenite

```python
def convert_temp(temp,sign):
    if sign=='C':
        return temp*9/5 + 32  ##celcius to ferenhite
    elif sign=='F':
        return (temp-32)*5/9
    else:
        return None
```

```python
print(convert_temp(37,'C'))
print(convert_temp(37,'F'))
print(convert_temp(56,'A'))
```

PASSWORD STRENGTH CHECKER


```python
def isstrong(password):
    if len(password)<8:
        return False
    if not any(char.isdigit() for char in password):
        return False
    if not any(char.islower() for char in password):
        return False
    if not any(char.isupper() for char in password):
        return False
    if not any(char in '!@#$%&*+_-?><,./' for char in password):
        return False
    else:
        return True
    
```

```python
print(isstrong('@MitabH08'))
print(isstrong('08Narayan'))
```

FACTORIAL OF NUMBER

```python
 def fact(num):
    if num==0 or num==1:
        return 1
    else:
        return num*fact(num-1)
```

```python
print(fact(5))
```

A FUNCTION TO READ A FILE AND COUNT THR FREQUENCY OF WORD


```python
def count_word(file_path):
    word_count = {}

    with open(file_path, 'r') as file:
        for line in file:
            words = line.split()
            for word in words:
                word = word.lower().strip('.,!?;:|"')
                word_count[word] = word_count.get(word, 0) + 1

    return word_count


filepath = 'sample.txt'
print(count_word(filepath))
```
